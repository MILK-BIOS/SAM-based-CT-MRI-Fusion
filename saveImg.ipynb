```python
import cv2
import os
import matplotlib.pyplot as plt
```

```python
sliceimg = 14
```

```python
cap = cv2.VideoCapture('dataset/Cerebrovascular_Disease/Hypertensive_encephalopathy_1/CT/spatial (1).mpg')
n = 0
while cap.isOpened():
    ret, frame = cap.read()
    n += 1
    if not ret or n == sliceimg:
        break
print(n)
CT = cap.read()[1]
plt.imshow(CT)
plt.axis('off')
CT = cv2.cvtColor(CT, cv2.COLOR_BGR2GRAY)
print(CT.shape)
```

```python
cap = cv2.VideoCapture('dataset/Cerebrovascular_Disease/Hypertensive_encephalopathy_1/MR-T2/spatial (1).mpg')
n = 0
while cap.isOpened():
    ret, frame = cap.read()
    n += 1
    if not ret or n == sliceimg:
        break
print(n)
MRI = cap.read()[1]
plt.imshow(MRI)
plt.axis('off')
MRI = cv2.cvtColor(MRI, cv2.COLOR_BGR2GRAY)
print(MRI.shape)
```

```python
save_dir = './testImg'
CT_path = os.path.join(save_dir, 'CT', f'{sliceimg}.jpg')
MRI_path = os.path.join(save_dir, 'MRI', f'{sliceimg}.jpg')
```

```python
cv2.imwrite(CT_path, CT)
cv2.imwrite(MRI_path, MRI)
```
